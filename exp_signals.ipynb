# Exponential signals

Generally, the exponential signals are expressed as 

$
\begin{align}
x(t)= A \exp(st) \;,
\end{align}
$

where $A$ and $s$ are complex constants. It is convenient to express a _complex amplitude_ $A$ in polar form 
$A=|A|e^{j\angle A}$ and an exponent constant $s$ in Cartesian form: $s=\sigma +j \omega$. 

Consider a simplest possible case where both $A$ and $s$ are real and $A=1$ to keep graphs on the same scale.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
s   = 0.4
tau = 1/s
print('Decay time = ', 1/s)
print('1/e   = ',1/e)
print('1/e^2 = ',1/e**2)
print('1/e^3 = ',1/e**3)

Ntau = 3  # Number of decay times to display
Points_per_tau = 5
t = np.linspace(0, Ntau*tau, Ntau*Points_per_tau)

x = np.exp( -s*t )
x1= np.exp( -2*s*t )
x2= np.exp( -0.5*s*t )
fig, exp_plot = plt.subplots()

# Horisontal grid linesare explicitely set at x = 1, 1/e, 1/e^2, 1/e^3 
exp_plot.set_yticks([0,1/e**3, 1/e**2, 1/e, 1], minor=False)
#Vertical grid lines are set at t= tau, 2*tau, 3*tau
exp_plot.set_xticks([0,tau, 2*tau,3*tau], minor=False)

exp_plot.yaxis.grid(True, which='major')
exp_plot.xaxis.grid(True, which='major')
plt.xlabel('time t')
plt.ylabel('x(t)')
plot(t, x, 'r')
plot(t, x1, 'b')
plot(t, x2, 'g')
plt.show()


```

## Interactive plot

First, declare imports to facilitate interactive features: ipywidgets

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
```

Define a module to plot an exponential signal. 1) Fixed interval in time $t_{\rm max}=3$.

```python
def exp_s(A, s):
    tmax = 3
    tau=1/s
    t = np.linspace(0, tmax, 100)
    x= A*np.exp( -s*t )
    fig, exp_plot = plt.subplots()
    plt.plot(t, x)
    plt.ylim(0, A)
    plt.xlim(0, tmax)
    exp_plot.set_yticks([0,A/e**3, A/e**2, A/e, A], minor=False)
    exp_plot.yaxis.grid(True, which='major')
    exp_plot.set_xticks([0,tau, 2*tau,3*tau], minor=False)
    exp_plot.xaxis.grid(True, which='major')
    plt.xlabel('time t')
    plt.ylabel('x(t)')
    plt.show()
```

```python
interactive_plot = interactive(exp_s, A=(0.5, 2.0, 0.5), s=(0.5, 5, 0.5))
interactive_plot
```

## Universal plot

Another module to plot an exponential signal. 2) Fixed number of decay times: $\tau=1/s$; time axis limit is set to three decay times. Now both axes are scaled to charasteristic values: time axis to a fixed number of decay times and signal axis to a signal amplitude.

```python
def exp_tau(A, s):
    tau=1/s
    tmax = 3*tau
    t = np.linspace(0, tmax, 100)
    x= A*np.exp( -s*t )
    fig, exp_plot = plt.subplots()
    plt.plot(t, x)
    plt.ylim(0, A)
    plt.xlim(0, tmax)
    exp_plot.set_yticks([0,A/e**3, A/e**2, A/e, A], minor=False)
    exp_plot.yaxis.grid(True, which='major')
    exp_plot.set_xticks([0,tau, 2*tau,3*tau], minor=False)
    exp_plot.xaxis.grid(True, which='major')
    plt.xlabel('time t')
    plt.ylabel('x(t)')
    plt.show()
```

```python
interactive_plot = interactive(exp_tau, A=(0.5, 2.0, 0.5), s=(0.5, 5, 0.5))
interactive_plot
```
